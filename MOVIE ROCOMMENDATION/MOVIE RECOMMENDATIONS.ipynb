```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df1 = pd.read_csv("ADD FILE PATH", quoting = True)
df1.head()
```

```python
df2 = pd.read_csv("ADD FILE PATH")
df2.head()
```

```python
print(df1.isnull().sum())
print(" ")
print(df2.isnull().sum())
```

```python
df1.dropna(subset = ['homepage', 'tagline'], inplace = True)
print(df1.isnull().sum())
```

```python
df1 = df1.drop(['budget','homepage','original_title','original_language','popularity','production_companies','production_countries','release_date','revenue','runtime','spoken_languages','status','tagline','vote_average','vote_count'], axis=1)
```

```python
df1 = df1[['id', 'title', 'genres', 'overview', 'keywords']]
final = df1.merge(df2, left_on='id', right_on= 'movie_id')[['title_x','genres','keywords','overview','cast','crew']]
df1.head()
```

```python
final.shape
```

```python
import ast
x = final.iloc[0,1]
print(x)
ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
def clean_text(text):
    x = []
    for i in ast.literal_eval(text):
        x.append(i['name'])
    y = x[:]
    return(y) 
final['genres']=final['genres'].apply(clean_text)
final['keywords']=final['keywords'].apply(clean_text)

def clean_cast(text):
    x = []
    counter = 0
    for i in ast.literal_eval(text):
        x.append(i['name'])
        counter = counter + 1;
        if counter == 3:
            break
    y = x[:]
    return(y)
final['cast']=final['cast'].apply(clean_cast)
```

```python
final.shape
```

```python
def clean_crew(text):
    x = []
    for i in ast.literal_eval(text):
        if i['job'] == "Director":
            x.append(i['name'])
    y = x[:]
    return(y) 
final['crew']=final['crew'].apply(clean_crew)
```

```python
final = final.rename(columns={"crew":"director"})
print(final.shape)
print(" ")
print(final.iloc[233,3])
```

```python
def remove_spaces(L):
    x=[]
    for i in L:
        x.append(i.replace(" ",""))
    y = x[:]
    return y
final['genres']= final['genres'].apply(remove_spaces)
final['cast']= final['cast'].apply(remove_spaces)
final['director']= final['director'].apply(remove_spaces)
final['keywords']= final['keywords'].apply(remove_spaces)
final['overview']= final['overview'].apply(remove_spaces)
```

```python
print(final.iloc[233,3])
```

```python
final['final_data']= final['genres']+final['keywords']+final['overview'] + final['cast'] + final['director']
final= final[['title_x','final_data']]
final.head()
```

```python
final.iloc[0,1]
```

```python
def list_to_string(L):
    str=" "
    return (str.join(L))
    
final['final_data'] = final['final_data'].apply(list_to_string)
final.head()
```

```python
final.iloc[0,1]
```

```python
final['final_data'] = final['final_data'].apply(lambda x:x.lower())
final.head()
```

```python
final.iloc[0,1]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv= CountVectorizer(stop_words='english', max_features=5000 )
X = cv.fit_transform(final['final_data']).toarray()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity_matrix = cosine_similarity(X)
```

```python
def recommend(movie):
    # recommend closest recommendations
    movie_index = final[final['title_x'] == movie].index[0]
    similarity_scores = similarity_matrix[movie_index]
    values = sorted(list(enumerate(similarity_scores)),reverse=True,key=lambda x:x[1])
    count = 0
    for i in values:
        count+=1
        if(count == 26):
            break
        print(final.iloc[i[0]].title_x) 
```

```python
recommend("Avengers: Age of Ultron")
```
